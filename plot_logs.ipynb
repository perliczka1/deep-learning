```python
%matplotlib inline
from matplotlib import pyplot
import numpy as np
```

```python
def plot_logs(path, epoch_from=0):
    # read data
    data =  np.loadtxt(path, delimiter=';')
    data = data[data[:,0]>=epoch_from]
    # plot loss
    fig = pyplot.figure(figsize = (23,7))
    fig.gca().grid(True)
    pyplot.plot(data[:,0], data[:,1])
    pyplot.title('Training loss')
    #plot accuracy
    fig = pyplot.figure(figsize = (23,7))
    fig.gca().grid(True)
    data = data[np.logical_and(~np.isnan(data[:,2]),~np.isnan(data[:,3]))]
    pyplot.plot(data[:,0], data[:,2], color = 'pink')
    pyplot.plot(data[:,0], data[:,3],  color = 'r')
    pyplot.legend(['Training batches', 'Validation'], loc=4)
    pyplot.title('Accuracy')
    pyplot.xlabel('Epoch')
```

```python
path = 'run3/96.08_ne.50_bs.128_lr.0.001_d.0.75_l2.0.006_t.csv'
plot_logs(path,1)
```
